## **Loading an Image**

To load an image, we can use the following syntax:
```python
image = cv2.imread(<filename>, <flag>)
```

- `filename` is the full path of the image file.
- `flag` is a parameter that specifies how the image should be read (RGB, Grayscale, etc.).
- Fot the `flag` parameter
    - $0$ means the image is read as grayscale
    - $1$ means the image is read as RGB
    - $-1$ means the image is read as RGBA (including alpha channel)

For example:
```
image = cv2.imread("assets/apel.png", 0)
```
In the code above, we read an image located in the `assets` folder with the filename `apel.png`, and it is loaded as a **grayscale** image.

## **Display an Image**

To display an image, we can use several methods:
1. Using the OpenCV GUI function `cv2.imshow()`
2. Using `plt.imshow()` from Matplotlib

## **OpenCV Way**

To display an image using OpenCV, we can use the `cv2.imshow()` syntax as shown below:

```python
cv2.imshow(<window_name>, <image>)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- `window_name` is the title of the GUI window.
- `image` is the image you want to display.
- `cv2.waitKey(<delay>)` decides for how long the image will be displayed with parameter `delay`. 
    - If `delay` $\leq 0$ the image will be shown forever.
    - Otherwise, the window will destroyed after delay milliseconds.

- `cv2.destroyAllWindows()` simply destroy all GUI windows created.

````python
cv2.imshow("???", "???")
cv2.waitKey(0)
cv2.destroyAllWindows()
````

## **Matplotlib Way**
To display an image using Matplotlib, we can use the `plt.imshow()` syntax as shown below:

```python
plt.imshow(<image>)
plt.show()
```

- `image` is the image you want to display.
- `plt.show()` is used to render and display the image directly inside the notebook.

## **Why the image looks different ?**

This happens due to the difference in color channel conventions between OpenCV and Matplotlib. OpenCV reads images in BGR order, while Matplotlib expects RGB order. Therefore, we need to swap the channels to display the correct colors.

We can do this using `cv2.cvtColor(<image>, <flag>)`:

- `image` is the image whose color channels you want to convert.
- `flag` specifies how the image should be converted (for example: `cv2.COLOR_BGR2RGB` converts from BGR to RGB).

````python
image_rgb = cv2.cvtColor("???", "???")
````

````python
plt.imshow(image_rgb)
plt.show()
````

## **Understanding the Image**

- Use `image.shape` to retrieve the dimensions of the image: `(height, width, channels)` for color images, or `(height, width)` for grayscale images.

- To get total number of pixels, we can use `image.size`, which returns the total number of elements `(height × width × channels)`.

- To access pixel intensity, we can use `image[y, x]` to get the intensity value at coordinates `(x, y)` (note that OpenCV uses row = y, column = x indexing).

- To access the data type, use `image.dtype` which return the pixels data type.
